```python
#importing library
import numpy as np
import pandas as pd
```

```python
#import dataset product 
ds=pd.read_csv("product_data.csv")
print(ds.to_string())
```

```python
#check basic description
ds.info() #check information about dataset
```

```python
#check duplicates rows
print(ds.duplicated())
print(ds.duplicated().sum())
```

```python
#drop duplicates column

ds=ds.drop_duplicates()
ds.info()
```

```python
#handle price column

ds["Price"]=pd.to_numeric(ds["Price"],errors="coerce")
```

```python
#handle null values in price column

ds["Price"]=ds["Price"].fillna(ds["Price"].mean())
print(ds.to_string())
```

```python
#Handle Quantity column
print("Values befor cnvort float :")
ds["Quantity"].unique()
ds["Quantity"]=pd.to_numeric(ds["Quantity"],errors="coerce")
ds["Quantity"].unique()
print(ds.to_string())
```

```python
#handling null cell in quantity column

ds["Quantity"]=ds["Quantity"].fillna("Null")
ds.info()
```

```python
#handling purchaseDate

ds["PurchaseDate"]=pd.to_datetime(ds["PurchaseDate"],errors="coerce")
ds.info()
```

```python
#save cleaned file

ds.to_csv("Clean_Product_dataset.csv")
print("dataset save successfully")
```
